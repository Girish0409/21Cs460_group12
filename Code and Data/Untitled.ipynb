```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.transform import rescale
from os import listdir
```

```python
# load all images in a directory
X = []
Y = []
for filename in listdir('C:/Users/DELL/Documents/GitHub/21Cs460_group12/Code and Data/WBC_data_kaggle/main_dataset/test/EOSINOPHIL'):
    # load image
    img_data = imread('C:/Users/DELL/Documents/GitHub/21Cs460_group12/Code and Data/WBC_data_kaggle/main_dataset/test/EOSINOPHIL/' + filename)
    #im_data = img_data[:,:,2]
    #Rescaling the image
    image_r = rescale(img_data, 0.4, anti_aliasing=True)
    X.append(image_r)
    Y.append(0)
```
